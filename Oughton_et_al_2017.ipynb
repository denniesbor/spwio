## Leontief

```python
# Identity Matrix
I = np.eye(A.shape[1])

# Leontief inverse
L = np.linalg.inv(I - A)
```

## Ghosh

Since we do not have 2017 gross output data, we will approximate industry allocation coefficient using 2021 gross output chained 2017 dollars

```python
# Ghosh calculation
# 2021 gross output
gross_output = pd.read_csv("/content/drive/MyDrive/spwio/data/us_gross_output.csv", usecols=["2021Q1"])

# Convert GDP vals as numpy array
x = gross_output["2021Q1"].to_numpy()

# Computing B matrix
gross_diag = np.diag(x) # Make the industrial output as the diagonal in a n by n matrix where n is number of industries/commodities
gross_diag_inv = np.linalg.inv(gross_diag) # Inverse of the diagonal matrix of outputs
B = np.dot(np.dot(gross_diag_inv, A), gross_diag) # Get the B coefficient matrix

print(B[1])
```

```python
def get_linkages(scenario_shocks: pd.DataFrame):

  """
  Calculate forward and backward linkages from direct shocks on gross value added using matrix decomposition.

  The approach is based on matrix decomposition methods proposed by Sonis et al., 1995.
  The function uses the input-output matrices A (technical coefficient matrix) and B
  (Leontief inverse matrix) to assess the indirect impacts of shocks across sectors.

  Parameters
  ----------
  scenario_shocks : pd.DataFrame

  Returns
  -------
  dict
    A dictionary containing two keys, 'shocks_FL' and 'shocks_BL', which map to arrays representing the
    calculated forward and backward linkages for each sector as a result of the direct shocks.

  """

  # Define the lists to collect results
  results = {
      "shocks_FL": [],
      "shocks_BL": [],
  }

  # Iterate through GDP shocks, assuming GDPShocks is defined elsewhere
  for s, row in enumerate(scenario_shocks.iterrows()):

    # Direct impact in sector s
    direct_impact = row[1]["2017"]

    print

    # Compute A* and A_s.
    A_star = np.delete(np.delete(A, s, axis=0), s, axis=1)
    A_s = np.delete(A[:, s], s)

    # Inverse term calculation
    inverse_I_A_star = np.linalg.inv(np.eye(A_star.shape[0]) - A_star)

    # Upstream impacts
    results["shocks_BL"].append(inverse_I_A_star @ A_s * direct_impact)

    # Calculate forward linkages using Sonis
    # Downstream impacts / Ghosh model
    B_star = np.delete(np.delete(B, s, axis=0), s, axis=1)
    B_s = np.delete(B[:, s], s)

    results["shocks_FL"].append(direct_impact * B_s @ np.linalg.inv(np.eye(B_star.shape[0]) - B_star))


  results["shocks_FL"] = np.array(results["shocks_FL"])
  results["shocks_BL"] = np.array(results["shocks_BL"])

  return results
```

```python
# Estimate the forward and backward linkages for scenario one
scenario_one_linkages = get_linkages(scenario_one_shocks)

# Print sum of linkages
print("Backward linkages: {} \n Forward Linkages {}" .format(scenario_one_linkages["shocks_BL"].sum(), scenario_one_linkages["shocks_FL"].sum()))
```

```python
# Estimate the forward and backward linkages for scenario two
scenario_two_linkages = get_linkages(scenario_two_shocks)

# Print sum of linkages
print("Backward linkages: {} \n Forward Linkages {}" .format(scenario_two_linkages["shocks_BL"].sum(), scenario_one_linkages["shocks_FL"].sum()))
```
